## Importing Modules

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
```

# Importing Dataset

```python
data=pd.read_csv('iris.csv')
data.head(5)
```

## Checking for Null Values

```python
data.isnull().sum()
```

```python
data.Species.value_counts()
```

## Feature Engineering

```python
data.Species.unique()
```

```python
dic={'Iris-setosa':0,    'Iris-versicolor':1,    'Iris-virginica':2}
data.Species = data.Species.map(dic)
```

```python
data.head(2)
```

```python
X=data.drop(['Id','Species'],axis=1)
Y=data['Species']
```

## HyperParameter Tuning

```python
param = {"max_depth": [3, None], 
              "max_features": [1,2,3,4,5,6,7,8,9], 
              "min_samples_leaf": [1,2,3,4,5,6,7,8,9], 
              "criterion": ["gini", "entropy"]} 
m=DecisionTreeClassifier()
mod_cv=GridSearchCV(m,param_grid=param,cv=10,scoring='accuracy')
mod_cv.fit(X,Y)
```

```python
mod_cv.best_estimator_
```

```python
mod_cv.best_score_
```

## Creating Model

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=3)
```

```python
model=DecisionTreeClassifier(criterion='entropy', max_depth=3, max_features=3,
                       min_samples_leaf=4)
model.fit(x_train,y_train)
```

# Vizulizing Decision Tree

```python
from sklearn import tree
```

```python
plt.figure(figsize=(13,10))
tree.plot_tree(model,feature_names=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm'],
              class_names=['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'],
              filled=True)
```

## Making Predictions

```python
pred=model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
print('Confusion Matrix :\n \n',confusion_matrix(y_test,pred))
```

```python
print('Accuracy Score : ',accuracy_score(y_test,pred))
```
